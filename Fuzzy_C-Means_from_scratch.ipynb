```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import skfuzzy as fuzz

# Step 1: Generate sample data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Step 2: Transpose data for skfuzzy
X_transposed = X.T

# Step 3: Apply Fuzzy C-Means
n_clusters = 3
fuzzifier = 2.0

cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    X_transposed, c=n_clusters, m=fuzzifier, error=0.005, maxiter=1000, init=None
)

# Step 4: Assign labels based on max membership
labels = np.argmax(u, axis=0)

# Step 5: Visualize results
colors = ['red', 'green', 'blue']
for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i+1}', color=colors[i], alpha=0.6)

plt.scatter(cntr[:, 0], cntr[:, 1], marker='x', s=200, c='black', label='Centers')
plt.title("Fuzzy C-Means Clustering")
plt.legend()
plt.show()

# Optional: Print Fuzzy Partition Coefficient (FPC)
print("Fuzzy Partition Coefficient (FPC):", fpc)


```
